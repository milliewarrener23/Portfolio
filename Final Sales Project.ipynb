```python
print(df.isnull().sum())
df.dropna(inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df)
print("\nData types:\n", df.dtypes)
```

```python
#calculate total sales and total profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
# Calculate total sales for each product
top_sales = df.groupby('Product Name')['Sales'].sum().reset_index()
```

```python
# Sort by total sales in descending order to find top-selling products
top_sales_sorted = top_sales.sort_values(by='Sales', ascending=False)
print(top_sales_sorted)
```

```python
# Get the top 10 best-selling products
top_10_sales = top_sales_sorted.head(10)
print(top_10_sales)
```

```python
#group by country and calculate total sales
sales_by_state = df.groupby('State')['Sales'].sum().reset_index()
print(sales_by_state)
```

```python
# Sort by total sales in descending order
sales_by_state_sorted = sales_by_state.sort_values(by='Sales', ascending=False)
print(sales_by_state_sorted)
```

```python
# Calculate the correlation between Sales and Profit
correlation = df['Sales'].corr(df['Profit'])
print(f"Correlation between Sales and Profit: {correlation}")
```

```python
# Group by category and calculate total sales
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
print(sales_by_category)
```

```python
# Sort by total sales (optional)
sales_by_category_sorted = sales_by_category.sort_values(by='Sales', ascending=False)
print(sales_by_category_sorted)
```

```python
# Bar chart for sales by product category
plt.bar(sales_by_category_sorted['Category'], sales_by_category_sorted['Sales'], color='skyblue')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Year-Month for grouping
df['Year-Month'] = df['Order Date'].dt.to_period('M')

# Group by Year-Month and calculate total sales
monthly_sales = df.groupby('Year-Month')['Sales'].sum().reset_index()

# Convert 'Year-Month' back to datetime for plotting
monthly_sales['Year-Month'] = monthly_sales['Year-Month'].dt.to_timestamp()
```

```python
# Plot the line graph
plt.plot(monthly_sales['Year-Month'], monthly_sales['Sales'], marker='o', color='blue')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Scatter plot
plt.scatter(df['Sales'], df['Profit'], color='blue', alpha=0.6, edgecolor='k')
plt.title('Scatter Plot of Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Pivot table for sales by region and category
heatmap_data = df.pivot_table(values='Sales', index='Region', columns='Category', aggfunc='sum', fill_value=0)

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu", cbar=True)

plt.title('Sales by Region and Product Category')
plt.xlabel('Product Category')
plt.ylabel('Region')
plt.tight_layout()

plt.show()
```

```python
# Resample data by month and sum sales
monthly_sales = df['Sales'].resample('ME').sum()

# Plot the time series of sales
monthly_sales.plot(figsize=(10, 6), title='Monthly Sales Over Time', marker='o')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```
